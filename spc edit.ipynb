```python
from datasets import load_dataset,get_dataset_config_names

# dataset = load_dataset("nsusemiehl/SciERC")
# print(get_dataset_config_names("zj/SciERC"))
# from datasets import load_dataset

dataset = load_dataset("zj88zj/SCIERC")
```

```python
labels = [
    "Material",
    "Method",
    "Metric",
    "Task",
    "Generic",
    "OtherScientificTerm"
]
```

```python
label_list = [
    "O",
    "B-Material","I-Material",
    "B-Method","I-Method",
    "B-Metric","I-Metric",
    "B-Task","I-Task",
    "B-Generic","I-Generic",
    "B-OtherScientificTerm",
    "I-OtherScientificTerm"
]

label2id = {
    l: i for i, l in enumerate(label_list)
}

id2label = {
    i: l for i, l in enumerate(label_list)
}
```

```python
def process_doc(example):

    tokens = []
    tags = []

    offset = 0

    for sent, ents in zip(
        example["sentences"],
        example["ner"]
    ):

        n = len(sent)

        sent_tags = ["O"] * n

        for start, end, label in ents:

            start -= offset
            end -= offset

            if start < 0 or end >= n:
                continue

            sent_tags[start] = f"B-{label}"

            for i in range(start + 1, end + 1):
                sent_tags[i] = f"I-{label}"

        tokens.extend(sent)
        tags.extend(sent_tags)

        offset += n

    return {
        "tokens": tokens,
        "ner_tags":
            [label2id[t] for t in tags]
    }
```

```python
processed = dataset.map(
    process_doc,
    remove_columns=
        dataset["train"].column_names
)
```

```python
print(processed["train"][0])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
    "allenai/scibert_scivocab_cased"
)
```

```python
def tokenize_and_align_labels(
    examples
):

    tokenized = tokenizer(
        examples["tokens"],
        truncation=True,
        is_split_into_words=True
    )

    labels = []

    for i, label in enumerate(
        examples["ner_tags"]
    ):

        word_ids = tokenized.word_ids(
            batch_index=i
        )

        previous = None
        label_ids = []

        for word_id in word_ids:

            if word_id is None:
                label_ids.append(-100)

            elif word_id != previous:
                label_ids.append(
                    label[word_id]
                )

            else:
                label_ids.append(-100)

            previous = word_id

        labels.append(label_ids)

    tokenized["labels"] = labels

    return tokenized
```

```python
tokenized_ds = processed.map(
    tokenize_and_align_labels,
    batched=True
)
```

```python
from transformers import (
    AutoModelForTokenClassification
)

model = (
    AutoModelForTokenClassification
    .from_pretrained(
        "allenai/scibert_scivocab_cased",
        num_labels=len(label_list),
        id2label=id2label,
        label2id=label2id
    )
)
```

```python
from transformers import (
    TrainingArguments,
    Trainer
)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

training_args = TrainingArguments(
    output_dir="./scientific_ner",
    report_to="none",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=5,
)
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(
    tokenizer=tokenizer
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator
)
```

```python
def search_abstract_NER(query,k=5):
  # query="deep learning for medical image analysis"
  query_embedding=embedding_model.encode([query])
  faiss.normalize_L2(query_embedding)
  D,I=index.search(query_embedding,k)
  for score,idx in zip(D[0],I[0]):
    print("\n")
    print("similarity score:-",score)
    print("TITLE:-", df.iloc[idx]['title'])
    print("ABSTRACT:-", df.iloc[idx]['abstract'][:500])
    entities = ner(df.iloc[idx]['abstract'])

    # for e in entities:
    #     print(
    #         e["word"],
    #         e["entity_group"],
    #         round(e["score"], 3)
    #     )
    nlp = spacy.blank("en")
    doc = nlp(df.iloc[idx]['abstract'])
    
    spans = []
    print("Entities Found:")
    for ent in entities:
        print(
        f"{ent['word']} --> "
        f"{ent['entity_group']}"
        )
    for ent in entities:
        span = doc.char_span(
            ent["start"],
            ent["end"],
            label=ent["entity_group"]
        )
    
        if span:
            spans.append(span)
    
    doc.ents = spans

    html=displacy.render(
        doc,
        style="ent",
        jupyter=False)
    
    display(HTML(html))
```
